```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv('Student_Marks.csv')
```

```python
df.columns = ['sinif','saat','puan']
```

```python
df.head(3)
```

```python
y=df[['puan']]
x=df[['sinif','saat']]
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.coef_ # coefficients in x1 and x2
```

```python
model.intercept_ # intercept point in x=0
```

```python
model.predict([[4,5]])
```

```python
4*1.86+ 5*5.39+ -7.45
```

```python
model.score(x,y) # accuracy
```
