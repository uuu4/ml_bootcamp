```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df = pd.read_csv("Student_Marks.csv")
```

```python
df.head(3)
```

```python
y = df[["Marks"]]
x = df[["number_courses","time_study"]]
```

```python
df.info()
```

```python
l = LinearRegression()
model = l.fit(x,y)
```

```python
model.predict([[3,2]])
```

```python
df["Marks"].max()
```

```python
model.score(x,y)
```
