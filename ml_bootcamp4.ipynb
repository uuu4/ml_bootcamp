```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

```python
df =pd.get_dummies(df,columns=['sex','smoker','region'],drop_first=True)
```

```python
y = df[['charges']]
x = df.drop(columns=['charges'])
```

```python
lm = LinearRegression()
model = lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[19,26,0,1,1,0,0,1]])
```

```python
df_error = pd.DataFrame()
```

```python
df_error['y'] = y
```

```python
df_error.head()
```

```python
y_pred = model.predict(x)
```

```python
df_error['pred'] = y_pred
```

```python
df_error.head(0)
```

```python
df_error['error'] = y-y_pred
```

```python
df_error.head(0)
```

```python
df_error['squared_error']=df_error['error']**2
```

```python
df_error['abs_error'] = np.abs(df_error['error'])
```

```python
df_error['percent_error'] = np.abs((y-y_pred)/y)
```

```python
df_error.mean()
```
